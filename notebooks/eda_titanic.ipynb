```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les donn√©es
df = pd.read_csv("data/titanic.csv", sep=";")
```

```python
# Aper√ßu des donn√©es
print("üîç Aper√ßu des premi√®res lignes :")
print(df.head())
```

```python
print("\nüìã Infos g√©n√©rales :")
print(df.info())
```

```python
print("\nüìä Statistiques descriptives :")
print(df.describe(include='all'))
```

```python
# üîé Valeurs manquantes
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title("Valeurs manquantes")
plt.show()
```

```python
# ‚úÖ R√©partition des survivants
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=df)
plt.title("R√©partition des survivants (0 = Non, 1 = Oui)")
plt.show()
```

```python
# üé´ R√©partition des classes
plt.figure(figsize=(6, 4))
sns.countplot(x='pclass', data=df)
plt.title("R√©partition des classes")
plt.show()
```

```python
# üë©‚Äçü¶± Survie selon le sexe
plt.figure(figsize=(8, 4))
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survie en fonction du sexe")
plt.show()
```

```python
# üéÇ Distribution des √¢ges
plt.figure(figsize=(8, 4))
df['age'].hist(bins=30)
plt.title("Distribution des √¢ges")
plt.xlabel("√Çge")
plt.ylabel("Nombre de passagers")
plt.show()
```

```python
# üì¶ R√©partition de l'√¢ge par classe
plt.figure(figsize=(8, 5))
sns.boxplot(x='pclass', y='age', data=df)
plt.title("√Çge par classe")
plt.show()
```

```python
# üß† Corr√©lation entre variables num√©riques
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Matrice de corr√©lation")
plt.show()
```
